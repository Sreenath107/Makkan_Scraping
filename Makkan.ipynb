```python
import time
import requests
from IPython.display import clear_output
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.makaan.com/hyderabad-residential-property/buy-property-in-hyderabad-city'
req = requests.get(url)
soup = BeautifulSoup(req.text,'html')
print(soup)
```

```python
Title=soup.findAll('div',attrs={'class':'title-line '})
Ti_tle=[i.text for i in Title]
Ti_tle
print(len(Ti_tle))
```

```python
Type=soup.findAll('span',attrs={'class' : 'seller-type'})
Ty_pe=[i.text for i in Type]
Ty_pe
print(len(Ty_pe))
```

```python
Loc=soup.findAll('span', attrs={'itemprop' : 'addressLocality'})
L_oc=[i.text for i in Loc]
L_oc
print(len(L_oc))
```

```python
Rate_sq=soup.findAll('td', attrs={'class':'lbl rate'})
Ra_te_sq=[i.text for i in Rate_sq]
Ra_te_sq
print(len(Ra_te_sq))
```

```python
Status=soup.findAll('tr', attrs={'class' : 'hcol w44'})
St_atus=[i.text for i in Status]
St_atus
print(len(St_atus))
```

```python
Size=soup.findAll('td',attrs={'class' : 'size'})
S_ize=[i.text for i in Size]
S_ize
print(len(S_ize))
```

```python
Name=soup.findAll('div',attrs={'class':'seller-info'})
Na_me=[i.text for i in Name]
Na_me
print(len(Na_me))
```

```python
Price=soup.findAll('div',attrs={'data-type':'price-link'})
Pr_ice=[i.text for i in Price]
print(len(Pr_ice))
```

```python
pd.DataFrame({'Title':Ti_tle,
              'Type':Ty_pe,
              'Location':L_oc,
              'Rate/Sq':Ra_te_sq,
              'Status':St_atus,
              'Sq-ft':S_ize,
              'Name':Na_me,
              'Price':Pr_ice})
```

```python
base_url='https://www.makaan.com/hyderabad-residential-property/buy-property-in-hyderabad-city?_=1624793766183'
```

```python
Ti=[]
Ty=[]
Lo=[]
Ra=[]
St=[]
Si=[]
Na=[]
Pr=[]
```

```python
for i in range(1,2):
  time.sleep(3)
```

```python
print(f'Going to scrape data form {i} page')
clear_output(wait=True)
url=base_url+str(i)
req = requests.get(url)
soup= BeautifulSoup(req.text,'html')
```

```python
Title=soup.findAll('div',attrs={'class':'title-line '})
for j in Title:
  Ti.append(j.text)
print(len(Ti))
```

```python
Type=soup.findAll('span',attrs={'class' : 'seller-type'})
for j in Type:
  Ty.append(j.text)
print(len(Ty))
```

```python
Loc=soup.findAll('span', attrs={'itemprop' : 'addressLocality'})
for j in Loc:
  Lo.append(j.text)
print(len(Lo))
```

```python
Status=soup.findAll('tr', attrs={'class' : 'hcol w44'})
for j in Status:
  St.append(j.text)
print(len(St))
```

```python
Size=soup.findAll('td',attrs={'class' : 'size'})
for j in Size:
  Si.append(j.text)
print(len(Si))
```

```python
Name=soup.findAll('div',attrs={'class' : 'seller-info'})
for j in Name:
  Na.append(j.text)
print(len(Na))
```

```python
Rate=soup.findAll('div',attrs={'data-type':'price-link'})
for j in Rate:
  Ra.append(j.text)
print(len(Ra))
```

```python
data=pd.DataFrame({'Title':Ti,
                   'Type':Ty,
                   'Location':Lo,
                   'Status':St,
                   'Sq-ft':Si,
                   'Name':Na,
                   'Rate':Ra
                   })
data
```

```python
data.to_csv('MakaanScrap.csv',index=False)
```
